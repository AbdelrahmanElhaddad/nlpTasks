```python
import pandas as pd
import nltk
```

```python
data = pd.read_csv("netflix_reviews.csv")
data.head()
```

```python
content = data.content
content
```

### Word tokenization

```python
# a sample of them
wsample = nltk.word_tokenize(content[0])
print(wsample)
ssample = nltk.sent_tokenize(content[899])
print(ssample)
```

```python
# tokenize the entire dataset
tokenized = list()
for row in content:
    tokenized.append(list(nltk.word_tokenize(row)))
tokenized
```

### Now for Stemming

```python
# Try the two kinds of stemming that we have
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()
# to show some differences
print(content[19])
print(p_stemmer.stem(content[19]))
```

```python
from nltk.stem.snowball import SnowballStemmer
s_stemmer = SnowballStemmer(language='english')
print(content[299])
print(s_stemmer.stem(content[299]))
```

```python
# Generalize this to the entire dataset
p_stemmed_list = list()
s_stemmed_list = list()

for row in content:
    p_stemmed_list.append(p_stemmer.stem(row))
    s_stemmed_list.append(s_stemmer.stem(row))

s_stemmed_list
```
